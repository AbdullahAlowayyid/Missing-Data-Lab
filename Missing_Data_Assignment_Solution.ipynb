### Do the following steps for each column with missing values.
* Identity the column.
* How much precentage the missing value represents?

```python
df['Employee_Name']
```

```python
print('Number of missing data of Employee names: ', df['Employee_Name'].isnull().sum())
null_vals = df['Employee_Name'].isnull().sum()
tot = df['Employee_Name'].count() + null_vals
perc = (null_vals / tot) * 100
print('The precentage the missing value represents: ', round(perc, 1),'%')
```

```python
# I prefer to replace the Nan with empty string as it can be rewritten later either from ID or positopn etc.
df['Employee_Name'].fillna('', inplace = True)

print('After handling: ')
df.isna().sum()
```

```python
# same will go with DOB and Sex as it only can be determined from the employee himself/herself

#DOB
print('Number of missing data of date of birth: ', df['DOB'].isnull().sum())
DOB_null_vals = df['DOB'].isnull().sum()
DOB_tot = df['DOB'].count() + DOB_null_vals
DOB_perc = (DOB_null_vals / DOB_tot) * 100
print('The precentage the missing value of date of birth represents: ', round(DOB_perc, 1),'%')

#Sex
print('Number of missing data of Sex: ', df['Sex'].isnull().sum())
Sex_null_vals = df['Sex'].isnull().sum()
Sex_tot = df['Sex'].count() + Sex_null_vals
Sex_perc = (Sex_null_vals / Sex_tot) * 100
print('The precentage the missing value of sex represents: ', round(Sex_perc, 1),'%')

#
df['DOB'].fillna('', inplace = True)
df['Sex'].fillna('', inplace = True)

#later in the machine learning maybe we can determine the sex from the name
print('After handling: ')
df.isna().sum()
```

```python
# so far as the absences are unknown, they will propably be considered 0 by the company 'I believe so'

print('Number of missing data of absences: ', df['Absences'].isnull().sum())
Absences_null_vals = df['Absences'].isnull().sum()
Absences_tot = df['Absences'].count() + Absences_null_vals
Absences_perc = (Absences_null_vals / Absences_tot) * 100
print('The precentage the missing value of absences represents: ', round(Absences_perc, 1),'%')
print('================================')

df['Absences'].fillna(0, inplace = True)

print('After handling: ')
df.isna().sum()
```

```python
# starting with the salary, I will be taking the mean of the employees salary of the same department

print('Number of missing data of salary: ', df['Salary'].isnull().sum())
Salary_null_vals = df['Salary'].isnull().sum()
Salary_tot = df['Salary'].count() + Salary_null_vals
Salary_perc = (Salary_null_vals / Salary_tot) * 100
print('The precentage the missing value of salary represents: ', round(Salary_perc, 1),'%')
print('================================')

# applying algorithm
department_avg_salary = df.groupby('Department')['Salary'].mean()
def replace_missing_salary(row):
    if pd.isnull(row['Salary']):
        return department_avg_salary[row['Department']]
    return row['Salary']

# now we aplly it for the column
df['Salary'] = df.apply(replace_missing_salary, axis = 1)
df
```

```python
# check
print(df.isna().sum())
print('================================')

# we notice that there is one null value remaining in the salary column
missing_salary_df = df[df['Salary'].isna()]
print(missing_salary_df)
print('================================')

# when we display the dataset for that row we can see that this is the only person who works in Executive Office
print(df['Department'].value_counts())
print('================================')

# so, I will handle this by giving the person the average of the salary
overall_avg_salary = df['Salary'].mean()

def replace_remaining_missing_salary(row):
    if pd.isnull(row['Salary']):
        return replace_remaining_missing_salary
    return row['Salary']


# now we apply it for the column
df['Salary'] = df.apply(replace_remaining_missing_salary, axis = 1)

# now let's check it out
print('After handling: ')
print(df.isna().sum())
print('================================')
```

```python
# Date of Termination
print('Number of missing data of date of termination: ', df['DateofTermination'].isnull().sum())
DateofTermination_null_vals = df['DateofTermination'].isnull().sum()
DateofTermination_tot = df['DateofTermination'].count() + DateofTermination_null_vals
DateofTermination_perc = (DateofTermination_null_vals / DateofTermination_tot) * 100
print('The precentage the missing value of date of termination represents: ', round(DateofTermination_perc, 1),'%')
print('================================')

# as we know, dates of hiring and termination data type is object
print('The data type for date of hiring: ', df['DateofHire'].dtype)
print('The data type for date of termination: ', df['DateofTermination'].dtype)
print('================================')

# we will turn their data types to date
# firstly, we must now how is the date represented
print('Date of Hire column: ')
print(df['DateofHire'])
print('================================')


print('Date of Termination column: ')
print(df['DateofTermination'])
print('================================')


# now we will turn them
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'])

# let's check
print('The data type for date of hiring after changing: ', df['DateofHire'].dtype)
print('The data type for date of termination after changing: ', df['DateofTermination'].dtype)
print('================================')

print('Date of Hire column after changing: ')
print(df['DateofHire'])
print('================================')


print('Date of Termination column after changing: ')
print(df['DateofTermination'])
print('================================')
```

```python
# in my opinion, applying forward filling for the termination date
#                         is the best option here as it is the most logical way to determine it 

# to make this, I will sort the dates in ascending order
df = df.sort_values(by = ['DateofHire', 'DateofTermination'])

# now apply forward filling method
df['DateofTermination'].fillna(method = "ffill", inplace = True)

# the first records is NaN so we will make backward filling for it
df['DateofTermination'].fillna(method = "bfill", inplace = True)

# now reset the dataset to its original sorting order
df.sort_index(inplace = True)

# now lets check the dataset
print('Dataset after handling: ')

print(df)
print('================================')



# date of termination null values now...
print('Null values After handling: ')
print(df.isna().sum())
print('================================')
```

```python
# Manager ID
print('Number of missing data of Managers ID: ', df['ManagerID'].isnull().sum())
ManagerID_null_vals = df['ManagerID'].isnull().sum()
ManagerID_tot = df['ManagerID'].count() + ManagerID_null_vals
ManagerID_perc = (ManagerID_null_vals / ManagerID_tot) * 100
print('The precentage the missing value of Managers ID represents: ', round(ManagerID_perc, 1),'%')
print('================================')

# while searching in the dataset, I found that mangers ID exist in other records
# I will make a dictionary for the managers with their IDs
managerDict = df.dropna(subset = ['ManagerID', 'ManagerName']).set_index('ManagerName')['ManagerID'].to_dict()

# now aplly it for the dataset
df['ManagerID'] = df.apply(lambda row: managerDict.get(row['ManagerName'], row['ManagerID']), axis = 1)

# now lets check the dataset
print('Dataset after handling: ')
print(df)
print('================================')



# ManagersID null values now...
print('Null values After handling: ')
print(df.isna().sum())
print('================================')
```
